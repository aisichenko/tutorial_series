# How to get your Github token

Go to [Github](github.com)

Go to your profile picture, then select **Settings**:



Go to **Developer settings** --> **Personal Access Tokens**

Click **Generate new token**. You can call it `OCPI_lab_<your_name>`

Save the token. Mine looks like this:

`55827dc7997b832419524004b527dde171c041f1`

**Save this token somewhere on your computer (write it down somewhere)**

---

To clone a repository called `test_repo` under the UCSB organization, you would write:
    
`git clone https://username:token@github.com/ucsb/test_repo`

where `username` is your Github username and the token is what we generated above.
